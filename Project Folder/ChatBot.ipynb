```python
import pandas as pd
import numpy as np
import tensorflow as tf
import re
```

```python
s=pd.read_csv("sherlock-dataset.csv")
```

```python
s.head()
```

```python
for i in range(0,100):
    print('#', i+1)
    print(s.Line[i])
    
```

```python

def Replace(text):
    text = text.lower()
    text = re.sub(r"\n", "",  text)
    text = re.sub(r"[-()]", "", text)
    text = re.sub(r"\.", " .", text)
    text = re.sub(r"\!", " !", text)
    text = re.sub(r"\?", " ?", text)
    text = re.sub(r"\,", " ,", text)
    text = re.sub(r"i'm", "i am", text)
    text = re.sub(r"he's", "he is", text)
    text = re.sub(r"she's", "she is", text)
    text = re.sub(r"it's", "it is", text)
    text = re.sub(r"that's", "that is", text)
    text = re.sub(r"what's", "what is", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"won't", "will not", text)
    text = re.sub(r"can't", "cannot", text)
    text = re.sub(r"n't", " not", text)
    text = re.sub(r"n'", "ng", text)
    text = re.sub(r"ohh", "oh", text)
    return text
```

```python
text = []
for l in s.Line:
    text.append(Replace(l))
```

```python
for i in range(0,100):
    print(text[i])
```

```python
maxi = 50
text_list = []
for l in text:
    if len(l.split()) <= maxi:
        text_list.append(l)
       
```

```python
vocabulary ={}
for l in text_list:
    for word in l.split():
        if word not in vocabulary:
            vocabulary[word]=1
        else:
            vocabulary[word]+=1
```
